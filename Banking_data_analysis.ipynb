```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\acer\OneDrive\Desktop\banking_data.csv')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
#Ques1.) Age distribution of clients in bank?
#using hist() function to define the plot graph 
plt.hist(df['age'],color='Blue',edgecolor= 'black')
plt.xlabel('Client Age')                    #Defining X-axis 
plt.ylabel('Frequency')                     #Defining Y-axis 
plt.title('Age Distribution of Clients')    #Defining Title
plt.show()                                  #To show the plot graph 
```

```python
#Ques2.) How does the job vary among the clients ? 
print("\n\nThe type of job of the clients in the bank are given below:")
job_type = df['job'].unique()
#Using for loop to print every job type  
for i in job_type:
    print(i)
```

```python
#Ques3.) What is the marital status distribution of the clients ?
#using the value_counts() function to calculate the marital status 
mar_stat = df['marital_status'].value_counts()
print("\n\nThe marital status of the clients is given below:")
#To print the marital status of the clients 
print(mar_stat)
```

```python
#Ques4.) What is the level of education among the clients ?
print("\n\nThe level of education of the clients are given below:")
#using unique() function to get the levels of education 
level_edu = df['education'].unique()
#Using for loop to print the level of education 
for i in level_edu:
     print(i)
```

```python
#Ques5.) What proportion of clients have credit in default ?
print("\n\nThe proportion of clients having credit in default is given below:")
#value_counts() function is used to calculate the proportion of credits 
default=df['default'].value_counts('yes')
# Printing the value 
print(default)
```

```python
#Ques6.) What is the average distribution of average yearly balance among the clients?
#Using the hist() function calculate the average distribution 
plt.hist(df['balance'],color= 'green', edgecolor='black')
plt.xlabel('Average yearly balance')
plt.ylabel('Frequency')
plt.title("Distribution of average yearly balance")
plt.show()
```

```python
#Ques7.) How many clients have housing loans?
#using the value_counts() function to get the number of clients with loan
h_loan = df['housing'].value_counts()
print("\n\nThe data below shows the clients that have housing loans.")
# To print the value 
print(h_loan)
```

```python
#Ques8.) How many clients have personal loans?
#using the value_counts() function to get the number of clients with Per..loan
p_loan= df['loan'].value_counts()
print("\n\nThe data given below shows the number of clients having personal loan.")
# To print the value 
print(p_loan)
```

```python
#Ques9.) What are the communication types used for contacting clients during the campaign?
print("\n\nThe type of communication used to contact the clients during a campaign is given below:")
#Using the unique() function 
comm_types = df['contact'].unique()
#Using for loop to print the values one by one 
for i in comm_types:
    print(i)
```

```python
#Ques10.) What is the distribution of the last contact day of the month?
#Using the hist() function 
plt.hist(df['day'],color= 'pink', edgecolor='black')
plt.xlabel('last contact day of month')
plt.ylabel('Frequency')
plt.title("Distribution of last contact day of the month")
#Using the show() function to plot the graph 
plt.show() 
```

```python
#Ques11.) How does the last contact month vary among the clients?
print("\n\nThe data of client's last contact month is given below:")
#using the value_counts() function 
vary = df['month'].value_counts()
#to print the value 
print(vary)
```

```python
#Ques12.) What is the distribution of the duration of the last contact?
#Using the hist() function
plt.hist(df['duration'],color='indigo',edgecolor='black')
plt.xlabel('Duration of the last contact')
plt.ylabel('Frequency')
plt.title('Distribution of duration of last contact in seconds')
#Using the show() function to plot the graph 
plt.show()
```

```python
#Ques13.) How many contacts were performed during the campaign for each client?
print("\n\nThe data for number of contact performed during the campaign is given below:")
#using the value_counts() function 
count_values = df['campaign'].value_counts()
#To print the value
print(count_values)
```

```python
#Ques14.) What is the distribution of the number of days passed since the client was last contacted from a previous campaign?
#Using the hist() function
plt.hist(df['pdays'],color= 'yellow', edgecolor='black')
plt.xlabel('Number of days passed')
plt.ylabel('Frequency')
plt.title("Distribution of Number of days passed")
#Using the show() function to plot the graph 
plt.show()
```

```python
#Ques.15.) How many contacts were performed before the current campaign for each client?
#using the value_counts() function 
v_count= df['previous'].value_counts()
print("\n\nThe data given below shows the contacts performed before the current campaign for each client.")
#To print the value
print(v_count)
```

```python
#Ques16.) What were the outcomes of the previous marketing campaigns?
print("\n\nThe data for the outcomes of previous campaign is given below:")
#using the value_counts() function to get the outcomes
outcome = df['poutcome'].value_counts()
#To print the outcomes
print(outcome)
```

```python
#Ques17.) What is the distribution of clients who subscribed to a term deposit vs. those who did not?
#Using the hist() function
plt.hist(df['y'],color= 'cyan', edgecolor='black')
plt.xlabel('Subcription of term deposit')
plt.ylabel('Frequency')
plt.title("Distribution of Clients that have subscribed to a term deposit")
#Using the show() function to plot the graph 
plt.show()
```

```python
#Ques18.) Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?
numeric_df = df.select_dtypes(include=['int64','float64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot = True,cmap='PuBuGn',fmt=".2f")
plt.title("Correlation matrix of banking dataset.")
```
